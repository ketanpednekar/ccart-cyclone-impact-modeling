```python
# ðŸŸ¢ Step 1: Setup
%matplotlib inline
import os, sys, numpy as np, matplotlib.pyplot as plt, pandas as pd

base_path = r"C:\CMIP data\GitHub Clean\ccart-cyclone-impact-modeling"
scripts_path = os.path.join(base_path, "scripts")
sys.path.append(scripts_path)

# Import CCART modules
from load_historical_tracks import load_historical_tracks
from cluster_tracks import cluster_tracks_by_path
from refine_analogs_pca import refine_bhola_analogs
from apply_climate_modifiers import apply_climate_modifiers
from function_run_ccart_pipeline import run_ccart_pipeline

boundary_path = r"C:\CMIP data\cmip6\Climada\Country Shape Files\ne_10m_admin_0_countries\ne_10m_admin_0_countries.shp"
countries = ["India", "Bangladesh"]

# ðŸŸ¢ Step 2: Load Historical Tracks
tracks = load_historical_tracks(basin="NI", year_range=(1980, 2020), provider="usa", min_wind=40)

# ðŸŸ¢ Step 3: Cluster Tracks
labels = cluster_tracks_by_path(tracks, eps=1.0, min_samples=3)

# ðŸŸ¢ Step 4: Visualize Clusters
plt.figure(figsize=(10, 8))
for i, track in enumerate(tracks):
    lat, lon = track["lat"].values, track["lon"].values
    label = labels[i]
    color = f"C{label % 10}" if label != -1 else "gray"
    plt.plot(lon, lat, color=color, label=f"Cluster {label}" if label != -1 else "Noise", alpha=0.6)
handles, labels_plot = plt.gca().get_legend_handles_labels()
plt.legend(dict(zip(labels_plot, handles)).values(), dict(zip(labels_plot, handles)).keys(), loc="upper left")
plt.xlabel("Longitude"); plt.ylabel("Latitude")
plt.title("Cyclone Track Clusters (DBSCAN)")
plt.grid(True); plt.tight_layout(); plt.show()

# ðŸŸ¢ Step 5: Extract Bhola-like Cluster
target_cluster = 2
bhola_like_tracks = [t for i, t in enumerate(tracks) if labels[i] == target_cluster]
print(f"âœ… Extracted {len(bhola_like_tracks)} Bhola-like tracks from Cluster {target_cluster}")

# ðŸŸ¢ Step 6: Refine Using PCA + Cosine Similarity
refined_bhola_like_tracks = refine_bhola_analogs(tracks, labels, target_cluster=target_cluster, top_n=5)

# ðŸŸ¢ Step 7: Plot Refined Tracks
plt.figure(figsize=(10, 8))
for track in refined_bhola_like_tracks:
    lat, lon = track["lat"].values, track["lon"].values
    plt.plot(lon, lat, label=track.attrs.get("name", "UNNAMED"), alpha=0.6)
plt.xlabel("Longitude"); plt.ylabel("Latitude")
plt.title("PCA-Refined Bhola-like Tracks")
plt.grid(True); plt.legend(loc="upper left"); plt.tight_layout(); plt.show()

# ðŸŸ¢ Step 8: Run Baseline Impact for Each Refined Track
impact_summaries = []
for i, track in enumerate(refined_bhola_like_tracks):
    name = track.attrs.get("name", f"bhola_candidate_{i}")
    result = run_ccart_pipeline(name=name, year=2035, basin="NI", countries=countries,
                                 boundary_path=boundary_path, save_fig=False, track_override=track)

    # âœ… Fix 1: Graceful unpacking
    try:
        gdf_litpop, gdf_impact, gdf_track, track_xr = result
    except ValueError:
        gdf_litpop, gdf_impact, gdf_track = result
        track_xr = None

    total_impact = gdf_impact["impact"].sum() if result and not gdf_impact.empty else 0
    impact_summaries.append({"index": i, "name": name, "total_impact": total_impact, "track": track})

    # ðŸ§ª Step 8.1: Debug Block â€” Wind Stats, Hazard Matrix, Exposure Overlap
    max_wind = track["max_sustained_wind"].max().item()
    mean_wind = track["max_sustained_wind"].mean().item()
    print(f"ðŸŒ¬ï¸ Wind stats for {name}: Max = {max_wind:.1f} kn | Mean = {mean_wind:.1f} kn")

    from function_load_tc_tracks import generate_hazard
    hazard = generate_hazard(track, buffer_deg=3.0)
    print(f"ðŸŒ€ Hazard shape: {hazard.intensity.shape}")
    print(f"ðŸŒ€ Max hazard intensity: {hazard.intensity.max().item():.1f} kn")

    if result:
        from function_plot_diagnostics import plot_tc_impact
        plot_tc_impact(
            track=track,
            gdf_litpop=gdf_litpop,
            gdf_impact=gdf_impact,
            title=f"{name} â€“ Wind Field vs Exposure",
            boundary_gdf=None,
            save_fig=False
        )

# ðŸŸ¢ Step 9: Select Highest Impact Track
max_impact_candidate = max(impact_summaries, key=lambda x: x["total_impact"])
selected_track = max_impact_candidate["track"]
selected_name = max_impact_candidate["name"]
print(f"\nâœ… Selected storm: {selected_name} with total impact ${max_impact_candidate['total_impact']:,.0f}")

# ðŸŸ¢ Step 10: Apply Warming Modifiers
scenario_modifiers = {
    "SSP2-4.5": {"wind_boost": 1.08, "rmw_shrink": 0.90},
    "SSP3-7.0": {"wind_boost": 1.12, "rmw_shrink": 0.85},
    "SSP5-8.5": {"wind_boost": 1.18, "rmw_shrink": 0.80}
}
synthetic_selected_by_scenario = {
    scenario: apply_climate_modifiers(selected_track, **params)
    for scenario, params in scenario_modifiers.items()
}

# ðŸŸ¢ Step 11: Plot Synthetic Tracks
plt.figure(figsize=(10, 8))
colors = {"SSP2-4.5": "blue", "SSP3-7.0": "orange", "SSP5-8.5": "red"}
for scenario, track in synthetic_selected_by_scenario.items():
    lat, lon = track["lat"].values, track["lon"].values
    plt.plot(lon, lat, label=scenario, color=colors[scenario], alpha=0.7)
plt.xlabel("Longitude"); plt.ylabel("Latitude")
plt.title(f"{selected_name} â€” Synthetic Tracks Across SSP Scenarios")
plt.grid(True); plt.legend(loc="upper left"); plt.tight_layout(); plt.show()

# ðŸŸ¢ Step 12: Run CCART Diagnostics for Each Scenario
impact_results = {}
for scenario, track in synthetic_selected_by_scenario.items():
    result = run_ccart_pipeline(name=f"{selected_name}_{scenario.replace('.', '')}", year=2035,
                                 basin="NI", countries=countries, boundary_path=boundary_path,
                                 save_fig=True, track_override=track)
    impact_results[scenario] = result

# ðŸŸ¢ Step 13: Inspect Impact Outputs
for scenario, result in impact_results.items():
    if result:
        gdf_litpop, gdf_impact, gdf_track, track = result
        print(f"\nâœ… {scenario} Impact Summary:")
        print("ðŸ§® Impact zones:"); display(gdf_impact.head())
        print("ðŸ™ï¸ Exposure points:"); display(gdf_litpop.head())
        print("ðŸŒ€ Storm track:"); display(gdf_track.head())
    else:
        print(f"âš ï¸ {scenario} pipeline failed.")
```
